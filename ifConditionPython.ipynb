# If condition ,if else ,else if and nested if 

```python
#if False
if False:
    Print("Hello world")

print("only this")
```

```python
#if True
if True:
    print("Hello world!")
    
```

```python
#if
a=2%4
print(a)
if a==0:
    print("Hello")
    
```

```python
#if if 
if True:
    print("hello")

if False:
    print("Name")
```

```python
#if else
if True:
    print("hello")

else:
    print("Name")
```

```python
#If elif
if False:
    print("hello")

elif True:
    print("Name")
```

```python
#Nested if
if True:
    print("hello")
    if True:
        print("Thirupatho")
```
